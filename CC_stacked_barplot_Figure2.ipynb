```python
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        #print(patch.get_x())
        patch.set_x(patch.get_x() + diff * 0.5)
        #patch.set_x(0,1,2)

```

```python
# set plot style: grey grid in the background:


# set the figure size
sns.set(font_scale = 2)
sns.set_style("white")
#plt.rcParams['patch.edgecolor'] = 'none'
plt.figure(figsize=(5, 8))
plt.rcParams.update({'font.size': 28,'font.family':'sans-serif'})


palette = {
    'Variability':"#F5C2B7"
}

# bar chart 1 -> top bars (group of 'increase with age')
bar1 = sns.barplot(x="Reason",  y="Value", hue='Reason',palette=palette, data=df.loc[(df['Category']=='All') & (df['Direction with age']=='Increase')& (df['Reason']=='Variability')], linewidth=2.5, edgecolor='#4D4D4D') #, hue= 'Color')
change_width(bar1, .25)


# bar chart 2 -> bottom bars (group of 'smoker=Yes')
bar2 = sns.barplot(x="Reason",  y="Value", hue='Reason', palette=palette, data=df.loc[(df['Category']=='All') & (df['Direction with age']=='Decrease')& (df['Reason']=='Variability')],linewidth=2.5,edgecolor='#94221F')
change_width(bar2, .25)
bar2.axhline(0, color='black', linewidth=2.5)

# add legend
top_bar = mpatches.Patch(color='#EC7063', label='Increase')
bottom_bar = mpatches.Patch(color='#4D4D4D', label='Decrease')

bar1.get_legend().remove()
#plt.legend(handles=[top_bar, bottom_bar],bbox_to_anchor=(1.05, 1.05))
plt.xlabel('')
plt.ylabel('Percentage of all genes')
plt.ylim(-100,100)
# show the graph
locs, labels = plt.xticks()
print(locs,labels)

plt.show()
```
